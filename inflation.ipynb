# Saudi Arabia Inflation Forecasting: A Univariate SARIMA Approach

This notebook develops a univariate seasonal time series forecasting framework to model and predict the **Consumer Price Index (CPI) / Inflation** (`inf`) for Saudi Arabia. The model focuses purely on the historical dependency and seasonal cycles of the inflation series itself, without external indicators.

## 📊 Model Specification

Through automated optimization via `pmdarima` (minimizing the Akaike Information Criterion - AIC), the optimal structural engine selected for the series is a pure **SARIMA** model:

$$\text{SARIMA}(0, 1, 0) \times (2, 0, 0)_{12}$$

### Parameter Deconstruction:
* **Non-Seasonal Order $(0, 1, 0)$:** * $p = 0$: Past consecutive monthly lags (lags 1–11) do not carry statistically significant predictive power once the trend is removed.
  * $d = 1$: The raw level inflation data is non-stationary and is stabilized using a single first-difference ($\Delta \text{INF}_t$), turning the target variable into month-on-month changes.
  * $q = 0$: No short-term moving average error shocks are required.
* **Seasonal Order $(2, 0, 0)_{12}$:**
  * $P = 2$: The model incorporates a second-order seasonal Autoregressive component. Current month-on-month inflation is regressed directly on its historical values from exactly 12 months ago (lag 12) and 24 months ago (lag 24).
  * $D = 0$: Seasonal differencing is unnecessary, as the combination of regular differencing ($d=1$) and seasonal AR terms completely handles the seasonal variance.
  * $s = 12$: Identifies the monthly cyclicality of the macroeconomic timeline.

---

## 📈 Mathematical Equation

The estimated state-space equation utilized to generate the in-sample fit and out-of-sample forecasts is defined as:

$$\Delta \text{INF}_t = c + \Phi_1 \Delta \text{INF}_{t-12} + \Phi_2 \Delta \text{INF}_{t-24} + \varepsilon_t$$

Where:
* $\Delta \text{INF}_t$ is the first difference of inflation at month $t$ ($\text{INF}_t - \text{INF}_{t-1}$).
* $c$ is the model intercept/constant.
* $\Phi_1$ is the coefficient for the 1-year-prior seasonal lag (lag 12).
* $\Phi_2$ is the coefficient for the 2-years-prior seasonal lag (lag 24).
* $\varepsilon_t$ is the white noise residual error term at time $t$.

## 🛠️ Environment & Core Workflow

This notebook executes the following analytical pipeline:
1. **Exploratory Data Analysis (EDA):** Visualizing historical inflation patterns and seasonal sub-series plots.
2. **Stationarity Testing:** Conducting Augmented Dickey-Fuller (ADF) unit root tests to verify the integration order $high(d=1)$.
3. **Identification:** Plotting Autocorrelation (ACF) and Partial Autocorrelation (PACF) functions to identify structural seasonal boundaries.
4. **Automated Search:** Implementing `pmdarima.auto_arima` stepwise search to isolate the optimal AIC parameters.
5. **Forecasting:** Generating future projections out of sample.

### Importing Libraries, inspecting and visualzing data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
pd.set_option('display.max_rows', 1000)
```

```python
df = pd.read_excel('Inflation.xlsx', sheet_name='Sheet1',index_col=0)
inf = df[['inf']]
inf.head()
```

```python
Inf = (df[['inf']]).diff()
Inf = Inf.dropna()
Inf.head()
```

```python
plt.figure(figsize=(12, 4),dpi=200)
plt.plot(inf.index, inf['inf'], color='tab:blue')
plt.title('Inflation Series (inf)')
plt.xlabel('Date')
plt.ylabel('inf')
#plt.grid(True)
plt.tight_layout()
plt.show()

adf_result = adfuller(inf['inf'].dropna(), autolag='AIC')
adf_stat, p_value, n_lags, n_obs, crit_values, icbest = adf_result

print(f"ADF Statistic: {adf_stat:.6f}")
print(f"p-value: {p_value:.6f}")
print(f"Used lags: {n_lags}")
print(f"Number of observations: {n_obs}")
print("Critical values:")
for key, value in crit_values.items():
    print(f"  {key}: {value:.6f}")

if p_value < 0.05:
    print("Interpretation: reject the null hypothesis of a unit root; the series appears stationary.")
else:
    print("Interpretation: fail to reject the null hypothesis of a unit root; the series appears non-stationary.")
```

```python
# Plot the differenced series and run ADF on it
plt.figure(figsize=(12, 4),dpi=200)
plt.plot(Inf.index, Inf['inf'], color='tab:blue')
plt.title('Differenced Inflation (Inf)')
plt.xlabel('Date')
plt.ylabel('diff inf')
#plt.grid(True)
plt.tight_layout()
plt.show()

adf_res = adfuller(Inf['inf'].dropna(), autolag='AIC')
adf_stat, p_value, used_lags, nobs, crit_vals, icbest = adf_res

print(f"ADF Statistic: {adf_stat:.6f}")
print(f"p-value: {p_value:.6f}")
print(f"Used lags: {used_lags}")
print(f"Number of observations: {nobs}")
print("Critical values:")
for k, v in crit_vals.items():
    print(f"  {k}: {v:.6f}")

if p_value < 0.05:
    print("Interpretation: reject the null hypothesis of a unit root; the differenced series appears stationary.")
else:
    print("Interpretation: fail to reject the null hypothesis of a unit root; the differenced series appears non-stationary.")
```

the time series is I(1) in levels as adf test shows

```python
import pmdarima as pm

y = inf.dropna()

model = pm.auto_arima(y,
                      start_p=0, max_p=12, start_q=0, max_q=12,
                      seasonal=True, m=12, d=None, trace=True,
                      information_criterion='aic',
                      error_action='ignore', suppress_warnings=True)
```

```python
print(model.summary())
```

```python
# Forecast 2026-04 to 2026-12 (9 months) using fitted `model`
n_periods = 9
preds, conf = model.predict(n_periods=n_periods, return_conf_int=True)

idx = pd.date_range(start=y.index[-1] + pd.offsets.MonthBegin(1), periods=n_periods, freq='MS')
df_forecast = pd.DataFrame({'forecast': preds, 'lower': conf[:, 0], 'upper': conf[:, 1]}, index=idx)

df_forecast
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df_forecast.index, df_forecast['forecast'], color='red', linestyle='-', marker='', linewidth=2)
plt.title('Forecasted Inflation')
plt.xlabel('Date')
plt.ylabel('Forecast')
plt.tight_layout()
plt.show()
```

# Conclusion & Key Takeaways

## 📌 Summary of Empirical Findings

This project successfully developed and evaluated a univariate **$\text{SARIMA}(0,1,0) \times (2,0,0)_{12}$** model to analyze and forecast monthly inflation (`inf`) in Saudi Arabia. By stripping away deterministic trends through first-differencing ($d=1$), the analysis isolated the structural stochastic dynamics of the series.

The model yields several critical macroeconomic insights:
1. **Absence of Short-Term Momentum ($p=0, q=0$):** The complete lack of non-seasonal autoregressive or moving average terms indicates that consecutive, month-on-month localized price shocks do not possess strong dynamic persistence. Random innovations in a given month are absorbed rapidly by the economy without triggering short-term inflationary spirals.
2. **Dominant Seasonal Memory ($P=2$):** Conversely, Saudi Arabia's inflation is characterized by a powerful year-over-year cyclical memory. The high statistical significance of the seasonal AR coefficients at lag 12 ($\Phi_1$) and lag 24 ($\Phi_2$) proves that current price behavior is heavily tethered to structural supply and demand dynamics from exactly one and two years prior. 

---

## 🔮 Forecast Implications (Target: 2026M04)

Because the model relies entirely on a second-order seasonal AR structure, the one-step-ahead forecast for **2026M04** is mathematically anchored to the historical data points of **2025M04** and **2024M04**. 

This pure univariate baseline provides a robust benchmark for policy evaluation. By capturing the underlying seasonal baseline, any significant divergence between this model's forecast and the actual flash estimates released by GASTAT can be confidently attributed to structural breaks, immediate domestic policy shifts, or unexpected global macroeconomic shocks rather than regular seasonal fluctuations.

---

## 🚀 Limitations & Future Structural Extensions

While the $\text{SARIMA}(0,1,0) \times (2,0,0)_{12}$ architecture serves as an exceptional univariate baseline, it operates under the assumption that historical cyclical patterns will continue uniformly. To build upon this framework, future research branches could investigate:
* **Structural Break Integration:** Utilizing intervention analysis (pulse dummy variables) to explicitly isolate known regulatory transformations, such as historical Value-Added Tax (VAT) implementations or major subsidy reforms.
* **Multivariate Expansion (SARIMAX):** Incorporating exogenous leading indicators, such as global wholesale food index changes or supply chain pressure indices, to map out the international pass-through transmission mechanisms affecting domestic retail prices.
